```python
#Pauli's Matrices
import numpy as np
import cmath as cm
sigma_x=np.array([[0,1],[1,0]])
sigma_y=np.array([[0,-1j],[1j,0]])
sigma_z=np.array([[1,0],[0,-1]])
sigma_x_theta=sigma_x.copy()
sigma_y_theta=sigma_y.copy()
sigma_z_theta=sigma_z.copy()
I=np.array([[1,0],[0,1]])
for i in range(2):
    for j in range(2):
        sigma_x_theta[i][j]=sigma_x[j][i].conjugate()
        sigma_y_theta[i][j]=sigma_y[j][i].conjugate()
        sigma_z_theta[i][j]=sigma_z[j][i].conjugate()
print("sigma_x= ",sigma_x, "\n","sigma_x_theta= ",sigma_x_theta,"\n","sigma_y= ",sigma_y, "\n","sigma_y_theta= ",sigma_y_theta,"\n","sigma_z= ",sigma_z, "\n","sigma_z_theta= ",sigma_z_theta)
def mat_multi(c,d):
    cd=np.array([[0,0],[0,0]])
    for i in range(2):
        for j in range(2):
            for k in range(2):
                cd+=c[i][k]*d[k][j]
def hermitian(a,b):
    for i in range(2):
        for j in range(2):
            if a[i][j]==b[i][j]:
                return True
if hermitian(sigma_x,sigma_x_theta):
    print("Hermitian Matrix")
if hermitian(sigma_y,sigma_y_theta):
    print("Hermitian Matrix")
if hermitian(sigma_z,sigma_z_theta):
    print("Hermitian Matrix")
```
