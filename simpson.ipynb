```python
def f(x):
    return 1/(1+x)
n=4
a=0
b=1
def simp(x):
    h=(b-a)/n
    I=(h/3)*(f(a)+f(b))
    for i in range(1,n):
        x=a+i*h
        if i%2==0:
            I+=(2*h/3)*f(x)
        if i%2!=0:
            I+=(4*h/3)*f(x)
    return I
for j in range(1,10):
    print(f"Integral={simp(0)}")
    n=2*n
print(f(a),f(b))
```
