```python
import random as rn
print("Random integers")
for i in range(3):
    r=rn.randint(1,16)
    print(r)
print("Binomial Variate")
for i in range(5):
    r=rn.binomialvariate(1,0.3)
    print(r)
print("Gauss Distribution")
for i in range(8):
    r=rn.gauss(0,1)
    print(r)
print("Uniform distribution")
for i in range(5):
    r=rn.uniform(0,1)
    print(r)
```
